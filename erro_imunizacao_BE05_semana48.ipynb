```python
df = df[['Nome Padrão']]

# Criando uma nova coluna para fazer a contagem dos erros
df['Total'] = 1

# Groupby = contagem somente de numeros 
df = df.groupby(['Nome Padrão']).sum()
df.reset_index(inplace=True)
df
```

```python
# Calcular taxa de incidencia

# Puxar banco das vacinas
df_vacinados = pd.read_csv('SESRS_VAC_APLICACAO_09_12_2021-10_33.csv', sep=';', encoding='cp1252')
df_vacinados

# filtrar data até 04-12-2021 (SE 48/2021) - CONFORME BOLETIM
df_vacinados = df_vacinados[df_vacinados['DATA']<= '2021-12-04']

# Total de vacinas aplicadas
total_aplicadas = df_vacinados['COUNT'].sum()
print(f'Total vacinados: {total_aplicadas}')
```

```python
# Calculo de incidencia por 100 mil

# Criando uma nova coluna para fazer a contagem dos erros
df['Incidencia'] = 1

for i in range(len(df['Total'])):
    df['Incidencia'][i] = "%.2f" % (df['Total'][i]/total_aplicadas*100000)
df
```

```python
# Ordenando os valores (maior para o menor)
df = df.sort_values('Incidencia', ascending=True)
df.to_excel('Erro imunizações.xlsx')
```

```python
df_final = pd.read_excel('Erro imunizações.xlsx')
df_final
```
